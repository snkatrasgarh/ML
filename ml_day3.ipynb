```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("/content/Position_Salaries.csv")
print(data)
```

```python
sns.heatmap(data.isnull())
```

```python
data=data.drop(['Position'],axis=1)
```

```python
print(data)
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
plt.scatter(x,y,color="blue")
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x,model.predict(x),color="red")
```

```python
from sklearn.preprocessing import PolynomialFeatures
myploy=PolynomialFeatures(degree=4)
poly_x=myploy.fit_transform(x)
print(poly_x)
```

```python
from sklearn.linear_model import LinearRegression
ploy_model=LinearRegression()
ploy_model.fit(poly_x,y)
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x,ploy_model.predict(poly_x),color="red")
```

```python
inp=[[3.5]]
poly_inp=myploy.fit_transform(inp)
yp=ploy_model.predict(poly_inp)
print(yp)
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x,ploy_model.predict(poly_x),color="red")
plt.scatter(inp,yp,color="black")
```
